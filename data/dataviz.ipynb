```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('data.csv')

#MAPPING
experience_mapping = {
    'SE': 'Senior',
    'MI': 'Mid Level',
    'EN': 'Entry Level',
    'EX': 'Executive'
}
employment_mapping = {
    'FT': 'Full Time',
    'PT': 'Partial Time',
    'CT': 'Casual',
    'FL': 'Freelancer'
}

#MAPPING
df['experience_level'] = df['experience_level'].map(experience_mapping)
df['employment_type'] = df['employment_type'].map(employment_mapping)

#GRAPH1
experience_counts = df['experience_level'].value_counts().reset_index()
experience_counts.columns = ['experience_level', 'count']
plt.figure(figsize=(8, 6))
sns.barplot(data=experience_counts, x='count', y='experience_level', palette='viridis', hue='experience_level', dodge=False, legend=False)
plt.title('Nível de Experiência')
plt.xlabel('Contagem')
plt.ylabel('Nível de Experiência')
plt.show()




```

```python
#GRAPH2
employment_counts = df['employment_type'].value_counts().reset_index()
employment_counts.columns = ['employment_type', 'count']
plt.figure(figsize=(8, 6))
sns.barplot(data=employment_counts, x='count', y='employment_type', palette='magma', hue='employment_type', dodge=False, legend=False)
plt.title('Distribuição por Tipo de Emprego')
plt.xlabel('Contagem')
plt.ylabel('Tipo de Emprego')
plt.show()
```

```python
# GRAPH3
top_locations = df['company_location'].value_counts().nlargest(10).reset_index()
top_locations.columns = ['company_location', 'count']
plt.figure(figsize=(10, 8))
sns.barplot(data=top_locations, x='count', y='company_location', palette='inferno', hue='company_location', dodge=False, legend=False)
plt.title('Top 10 Localizações de Empresas')
plt.xlabel('Contagem')
plt.ylabel('Localização da Empresa')
for i, v in enumerate(top_locations['count']):
    plt.text(v + 0.2, i, str(v), color='black')

plt.show()

```

```python
#GRAPH4
currency_counts = df['salary_currency'].value_counts().nlargest(5).reset_index()
currency_counts.columns = ['salary_currency', 'count']
currency_counts['percentage'] = (currency_counts['count'] / currency_counts['count'].sum()) * 100

plt.figure(figsize=(10, 6))
sns.barplot(data=currency_counts, x='percentage', y='salary_currency', palette='rocket', hue='salary_currency', dodge=False, legend=False)
plt.title('Distribuição das 5 Principais Moedas de Salário (em %)')
plt.xlabel('Porcentagem')
plt.ylabel('Moeda')

# Adicionando rótulos para cada barra
for i, v in enumerate(currency_counts['percentage']):
    plt.text(v + 0.2, i, f'{v:.2f}%', color='black')

plt.show()
```
